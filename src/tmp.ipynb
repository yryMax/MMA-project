```python
import os
path1 = '../imagefolders1/imagefolders/'
path2 = '../imagefolders2/imagefolders/'
#search for all files in the folder
files1 = os.listdir(path1)
files2 = os.listdir(path2)
files1.remove('.DS_Store')
pathnk = '../data/nk/'
pathoj = '../data/oj/'
pathrh = '../data/rh/'
pathothers = '../data/others/'
```

```python
for files in files2:
    for file in os.listdir(path2 + files):
        cat = file.split('_')
        print(cat)
        if len(cat) <= 2:
            continue
        cat = cat[2]
        print(cat)
        if cat == "nk":
            #move the file under pathnk
            os.rename(path2 + files + '/' + file, pathnk + file)
        elif cat == "oj":
            os.rename(path2 + files + '/' + file, pathoj + file)
        elif cat == "rh":
            os.rename(path2 + files + '/' + file, pathrh + file)
        else:
            os.rename(path2 + files + '/' + file, pathothers + file)
```
